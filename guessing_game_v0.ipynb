```python
import random

def play_game():
    
    attempts_limit = 3
    while True:
        try:
            print("Set range minimum:")
            min_range = int(input())
            if (min_range >=0):
                break
                
        except ValueError:
                print("Invalid range!")
                
    while True:    
        try:
            print("Set range maximum:")
            max_range = int(input())
        
            if (min_range < max_range):
                random_number = random.randint(min_range, max_range)
                print("Guess:")
                break
                    
        except ValueError:
            print("Invalid range!")
                
    while True:
        try:
    
            guess = int(input())
                
            if (guess > random_number):
                print ("Too high!")
                attempts_limit -= 1
                print(f"Attempts left: {attempts_limit}")
            elif (guess < random_number):
                print ("Too low!")
                attempts_limit -= 1
                print(f"Attempts left: {attempts_limit}")
            elif (guess == random_number):
                print("Great guess!")
                return
                
            if (attempts_limit == 0):
                print("No attempts left!")
                return
                    
        except ValueError:
            print("Invalid guess!")
            continue
            
playing = True 
while playing:
    play_game()
    print("Do you want to try again?")
    print("y/n")
    while True:
        restart = input().lower()
        if (restart == "y"):
            break
        elif (restart == "n"):
            print("Thx for playing!")
            playing = False
            break
        else:
            print("Invalid input")
            print("Do you want to try again?")
            print("y/n")
```
