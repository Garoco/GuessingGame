```python
import sys
import random
import math

class Game:

    def __init__(self):
        self.last_difficulty = None
        self.state = "menu"
        game_properties = {
            "menu": self.menu,
            "difficulty": self.difficulty,
            "easy": lambda: self.play_easy(1, 10),
            "medium": lambda: self.play_medium(1, 100, 5),
            "hard": lambda: self.play_hard(1, 1000, 3),
            "continue": self.game_end
        }

    def run(self):
        while self.state:
            self.state = game_properties[self.state]()
            
    def menu(self):
        print("Play\nExit")
        while True:
            menu_input = input().lower()
            if menu_input in {"play"}:
                return "difficulty"
            elif menu_input in {"exit"}:
                print("Thx for playing!")
                sys.exit()
            else:
                print("Invalid command!")
    
    def difficulty(self):
        print("Easy\nMedium\nHard\nReturn")
        
        while True:
            difficulty_input = input().lower()
            if difficulty_input in {"easy"}:
                self.last_difficulty = "easy"
                return "easy"
            elif difficulty_input in {"medium"}:
                self.last_difficulty = "medium"
                return "medium"
            elif difficulty_input in {"hard"}:
                self.last_difficulty = "hard"
                return "hard"
            elif difficulty_input in {"return"}:
                return "menu"
            else:
                print("Invalid choice!")
                
    def play_easy(self, min_range, max_range):
        random_number = random.randint(min_range, max_range)
        print("Guess!")
        while True:
            try:
                guess = int(input())
                
                if (guess > random_number):
                    print ("Too high!")
                elif (guess < random_number):
                    print ("Too low!")
                elif (guess == random_number):
                    print("Great guess!\nTry again?\nYes/No")
                    return "continue"
            except ValueError:
                print("Invalid guess!")

    def play_medium(self, min_range, max_range, attempts_limit):
            random_number = random.randint(min_range, max_range)
            print("Guess!")
            difference = None
            while True:
                try:
                    guess = int(input())
                    
                    if (guess > random_number):
                        print ("Too high!")
                        attempts_limit -= 1
                    elif (guess < random_number):
                        print ("Too low!")
                        attempts_limit -= 1
                    elif (guess == random_number):
                        print("Great guess!\nTry again?\nYes/No")
                        return "continue"

                    difference = abs(random_number - guess)

                    if (25 >= difference > 15):
                        print("Cold!")
                    elif (difference > 25):
                        print("Very cold!")
                    elif (15 >= difference > 10):
                        print("Warm!")
                    elif (10 >= difference > 5):
                        print("Very warm!")
                    elif (5 >= difference > 0):
                        print("Really hot!")
                    
                    if (attempts_limit > 0):
                        print(f"Attempts left: {attempts_limit}")
                        continue
                    elif (attempts_limit == 0):
                        print(f"No attempts left!\nThe number was {random_number}\nTry again?\nYes/No")
                        return "continue"
                        
                except ValueError:
                    print("Invalid guess!")

    def play_hard(self, min_range, max_range, attempts_limit):
            random_number = random.randint(min_range, max_range)
            print("Guess!")
            while True:
                try:
                    guess = int(input())
                    
                    if (guess > random_number):
                        print ("Too high!")
                        attempts_limit -= 1
                    elif (guess < random_number):
                        print ("Too low!")
                        attempts_limit -= 1
                    elif (guess == random_number):
                        print(f"Great guess!\nTry again?\nYes/No")
                        return "continue"

                    difference = abs(random_number - guess)

                    if (25 >= difference > 15):
                        print("Cold!")
                    elif (100 >= difference > 25):
                        print("Super cold!")
                    elif (difference > 100):
                        print("Extremely cold!")
                    elif (15 >= difference > 10):
                        print("Warm!")
                    elif (10 >= difference > 5):
                        print("Very warm!")
                    elif (5 >= difference > 0):
                        print("Really hot!")
                        
                    if (attempts_limit > 0):
                        print(f"Attempts left: {attempts_limit}")
                        continue
                    elif (attempts_limit == 0):
                        print(f"No attempts left!\nThe number was {random_number}\nTry again?\nYes/No")
                        return "continue"
                        
                except ValueError:
                    print("Invalid guess!")
                    
    def game_end(self):
        while True:
            choice = input().lower()
            if (choice == "yes"):
                return self.last_difficulty
            elif (choice == "no"):
                return "menu"
            else:
                print("Invalid command!")
                    
game = Game()
game.run()
```
